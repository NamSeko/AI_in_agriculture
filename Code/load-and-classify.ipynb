```python
import torchvision
import torch
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
classes = [
    'Bình thường',
    'Nấm đốm',
    'Phấn trắng',
]
```

```python
model = torch.load('/kaggle/input/serminar-khoa-hc/pytorch/model/1/best_model.pth')
```

```python
mean = [0.4363, 0.4328, 0.3291]
std = [0.2129, 0.2075, 0.2038]

image_transforms = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize(torch.Tensor(mean), torch.Tensor(std))
])
```

```python
def classify(model, image_transforms, image_path, classer):
    model = model.eval()
    image = Image.open(image_path)
    image = image_transforms(image).float()
    image = image.unsqueeze(0)
    
    output = model(image)
    _, predicted = torch.max(output.data, 1)
    
    predicted_label = classer[predicted.item()]

    # Hiển thị ảnh và nhãn
    plt.imshow(image.squeeze().permute(1, 2, 0))
    plt.title(predicted_label)
    plt.show()
```

## Check xem lá bị bệnh nấm đốm, phấn trắng hay bình thường

```python
pt = os.listdir("/kaggle/input/photo-of-infected-leaves/Test_data/PT")
nd = os.listdir("/kaggle/input/photo-of-infected-leaves/Test_data/ND")
bt = os.listdir("/kaggle/input/photo-of-infected-leaves/Test_data/BT")
```

### Check bệnh phấn trắng

```python
fig, axs = plt.subplots(5, 8, figsize=(12, 10))  # Adjust the figsize for desired image size

index = 0
for row in range(5):
    for col in range(8):
        if index < len(pt):
            image_path = "/kaggle/input/photo-of-infected-leaves/Test_data/PT/" + str(pt[index])
            model = model.eval()
            image = Image.open(image_path)
            image = image_transforms(image).float()
            image = image.unsqueeze(0)

            output = model(image)
            _, predicted = torch.max(output.data, 1)

            predicted_label = classes[predicted.item()]

            axs[row, col].imshow(image.squeeze().permute(1, 2, 0))
            axs[row, col].set_title(predicted_label)
            axs[row, col].axis('off')
        index += 1

plt.tight_layout()
plt.show()
```
